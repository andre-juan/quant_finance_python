## 1) Introdução

Todo investimento em ativos financeiros traz consigo **riscos**, que estão relacionados com às chances de ocorrerem **retornos negativos**, isto é, **perdas**.

No contexto de riscos, há duas grandes classes:

   - **risco não diversificável**: é aquele sobre o qual o investidor não tem controle, como incertezas políticas, oscilações da inflação ou taxas de juros. Trataremos sobre isso quando falarmos de CAPM;
   - **risco diversificável**: é aquele sobre o qual o investidor tem controle, e pode ser diminuído por meio da **distribuição do patrimônio em ativos com graus de risco diferentes**. Trataremos primeiro sobre como abordar este tipo de risco.
   
Ao investirmos, é muito importante que construamos um **portfolio de investimentos**, isto é, uma **carteira** composta por diferentes ativos.

Isto é muito importante devido à importância da **diversificação**: quanto mais diversificado um portfolio (no sentido que discutiremos), menores serão os **riscos** de observarmos perdas consideráveis.

## 2) Univariate investment risk

Começaremos nossa exposição abordando algo muito importante: como mensurar o risco de **um único ativo**, o que define o risco univariado.

**O que é risco?**

Quando falamos de riscos em investimentos financeiros, estamos nos referindo particularmente a:

- medida de incerteza;
- dispersão ou variância de retornos financeiros.

**Como medir risco?**

Há diversas formas de medirmos o risco de determinado ativo. As que veremos hoje são:

- Desvio padrão ou variância de retornos diários;
- Kurtosis da distribuição de retornos diária;
- Skewness da distribuição de retornos diária;
- Drawdown histórico;
- Value at Risk.

Vamos explorar cada um dos conceitos acima em python!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pandas_datareader.data import DataReader

%matplotlib inline
```

```python
########################################################################
################################# lendo os dados

simbolo = 'ITUB3.SA'

df = DataReader(simbolo, data_source='yahoo', start='2010')

# preços
precos = df["Adj Close"]

# retornos
retornos = precos.pct_change().dropna()

########################################################################
################################# plot dos preços

plt.figure(figsize=(14, 6))
plt.title(simbolo + " - série de precos")

sns.lineplot(x=precos.index, y=precos, label="ST de precos")

plt.legend()
plt.show()

########################################################################
################################# plot dos retornos

plt.figure(figsize=(14, 6))
plt.title(simbolo + " - série de retornos")

sns.lineplot(x=retornos.index, y=retornos, label="ST de retornos")

plt.legend()
plt.show()

########################################################################
################################# plot da distribuição dos retornos

plt.figure(figsize=(12, 6))

plt.title(simbolo + " - Distribuição dos retornos", size=18)

sns.histplot(retornos, kde=True, alpha=0.2, label="retornos")

plt.axvline(retornos.mean(), color="red", label="média")

plt.axvline(retornos.median(), color="green", label="mediana")

for i in range(retornos.mode().shape[0]):
    plt.axvline(retornos.mode()[i], color="yellow", label="moda")

plt.legend()
plt.show()
```

Primeiro passo importante é checar se os retornos são normais:

```python
from scipy.stats import normaltest
    
test_stat, p = normaltest(retornos)

print("Estatística de teste:", test_stat)
print("\np-value:", p)

# interpretação do teste
# nível de confiança de 5%
alpha = 0.05

if p > alpha:
    print("\nFalha em rejeitar H_0: amostra parece ser normal")
else:
    print("\nRejeição da H_0: amostra não é normal")
```

De fato, nossos retornos não são normais, conforme esperado. Assim, como nossa distribuição não é normalmente distribuída, devemos nos preocupar com retornos negativos muito comuns!

Em geral, muitas vezes nos encontramos com a seguinte situação:

- retornos financeiros tendem a não ter distribuição normal;
- retornos financeiros podem ter curtose alta;

Lembrando que:

- distribuições normais tem assimetria próxima de 0 e curtose próximo de 3.

Vamos agora olhar para algumas medidas de risco:

### 2.1) Volatilidade: desvio padrão e variância

Sempre que temos uma distribuição, é possível calcularmos a variância $\sigma^2$ e o desvio padrão (volatilidade) $\sigma$. 

Um ativo com alto $\sigma$ é visto como investimento de maior risco, dado que as incertezas e variações aumentam.

É importante lembrar que a volatilidade escala com a raiz quadrada do tempo. Podemos assumir 252 dias úteis no ano e 21 dias úteis no mês, assim:

$$\sigma_{ano} = \sigma_{dia}\sqrt{252}$$

$$\sigma_{mês} = \sigma_{dia}\sqrt{21}$$

```python
# volatilidade

print('\nVolatilidade diária do ativo', simbolo + " no período:", retornos.std(), '\n')

print('Volatilidade anualizada do ativo', simbolo + " no período:", retornos.std() * np.sqrt(252), '\n')
```

### 2.4) Max Drawdown

Uma medida de risco muito utilizada é o do **maximum drawdown (MDD)**, que quantifica qual foi a maior "queda" do preço de um ativo, ou seja, qual foi a maior diferença observada entre um pico e um vale.

Começamos calculando a **série de drawdowns**, que será, a cada dia, a diferença entre o rolling max e o próprio preço:

```python
plt.figure(figsize=(14, 6))
plt.title(simbolo + " - preço de fechamento ajustado")

# série original
sns.lineplot(x=precos.index, y=precos, label="ST de preços original")

# rolling max
rolling_max = precos.expanding().max()
sns.lineplot(x=rolling_max.index, y=rolling_max, label="Rolling max")

# rolling min
rolling_min = precos.expanding().min()
sns.lineplot(x=rolling_min.index, y=rolling_min, label="Rolling min")

plt.legend()
plt.show()
```

```python
precos
```

```python
rolling_max
```

```python
rolling_max = precos.expanding().max()

drawdowns = rolling_max - precos
```

```python
drawdowns
```

Uma vez calculada a série de drawdowns, basta tomar seu valor máximo para encontrar o max drawdown:

```python
drawdowns.max()
```

Para saber qual foi o dia do MDD:

```python
drawdowns.idxmax()
```

O valor acima é, portanto, a **diferença máxima de preços observada em todo o período**.

Podemos plotar os drawdowns:

```python
plt.figure(figsize=(14, 6))
plt.title(simbolo + " - Drawdowns")

sns.lineplot(x=drawdowns.index, y=drawdowns, label="ST de drawdowns")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.title(simbolo + " - preço de fechamento ajustado")

# série original
sns.lineplot(x=precos.index, y=precos, label="ST de preços original")

# rolling max
rolling_max = precos.expanding().max()
sns.lineplot(x=rolling_max.index, y=rolling_max, label="Rolling max")

# rolling min
rolling_min = precos.expanding().min()
sns.lineplot(x=rolling_min.index, y=rolling_min, label="Rolling min")

sns.lineplot(x=drawdowns.index, y=drawdowns, label="ST de drawdowns")

plt.legend()
plt.show()
```

```python
fig, (ax_preco, ax_dd) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.80, .20)}, figsize=(14, 6))

plt.suptitle(simbolo + " - preço de fechamento ajustado")

# plot dos preços
sns.lineplot(x=precos.index, y=precos, label="ST de preços original", ax=ax_preco)

rolling_max = precos.expanding().max()
sns.lineplot(x=rolling_max.index, y=rolling_max, label="Rolling max", ax=ax_preco)

rolling_min = precos.expanding().min()
sns.lineplot(x=rolling_min.index, y=rolling_min, label="Rolling min", ax=ax_preco)

# plot dos dds
mdd = drawdowns.max()
sns.lineplot(x=drawdowns.index, y=-drawdowns, 
             label="ST de drawdowns. MDD = " + str(round(mdd, 2)), ax=ax_dd, color="red")

# espaço entre os plots
fig.subplots_adjust(hspace=0)

plt.legend()
plt.show()
```

### 2.5) Value at Risk (VaR)

O **Value at Risk** faz parte do que é chamado de **tail risk**, ou seja, risco de cauda, pois ele estima **qual seria a perda do ativo em casos extremos**, ou seja, refere-se ao **risco de ocorrências de resultados extremos de investimento**, principalmente no lado negativo de uma distribuição (cauda da esquerda)**

O que o VaR calcula é **o valor a perda máxima possível em um horizonte de tempo, se excluindo as observações cuja probabilidade de observação seja no máximo $p$**.

Na prática, podemos aproximar o VaR ao **valor do p-quantil da distribuição de retornos**:

<img src="https://upload.wikimedia.org/wikipedia/commons/6/64/VaR_diagram.JPG" width=500>





```python
alpha = 0.05

retornos.quantile(alpha)
```

Ou seja, há 5% de chance de termos perdas maiores que o valor do VaR (ou seja, retornos menores que o valor do -VaR)

```python
plt.figure(figsize=(12, 6))

plt.title(simbolo + " - Distribuição dos retornos", size=18)

sns.histplot(retornos, kde=True, alpha=0.2, label="retornos")

plt.axvline(retornos.mean(), color="red", label="média")

plt.axvline(retornos.median(), color="green", label="mediana")

# plot com o var
alpha = 0.05
var = retornos.quantile(alpha)
plt.axvline(var, color="orange", label=str(alpha*100)+"% -VaR: " + str(round(var, 2)))

for i in range(retornos.mode().shape[0]):
    plt.axvline(retornos.mode()[i], color="yellow", label="moda")

plt.legend()
plt.show()
```

### 2.6) Conditional VaR (CVaR) ou Expected Shortfall

Esta medida nada mais é do que **o valor esperado da distribuição de retornos menores que o VaR**.

De maneira similar a o que fizemos com o VaR, aproximamos o CVaR pelo **retorno esperado nos $q\%$ piores casos**.

O CVaR também é uma medida de risco, alternativa ao VaR por ser menos sensível à forma da cauda.

<img src="https://i.ytimg.com/vi/lROJ5gX3BbQ/maxresdefault.jpg" width=600>

```python
var
```

```python
retornos
```

```python
cvar = retornos[retornos < var].mean()

cvar
```

```python
plt.figure(figsize=(12, 6))

plt.title(simbolo + " - Distribuição dos retornos", size=18)

sns.histplot(retornos, kde=True, alpha=0.2, label="retornos")

plt.axvline(retornos.mean(), color="red", label="média")

plt.axvline(retornos.median(), color="green", label="mediana")

# var
plt.axvline(var, color="orange", label=str(alpha*100)+"% -VaR: " + str(round(var, 2)))

# cvar
plt.axvline(cvar, color="blue", label=str(alpha*100)+"% -CVaR: " + str(round(cvar, 2)))

for i in range(retornos.mode().shape[0]):
    plt.axvline(retornos.mode()[i], color="yellow", label="moda")

plt.legend()
plt.show()
```

### 3.1) Retorno do portfolio

Um portfolio é composto por um conjunto de ativos, cada um presente em uma **proporção**, ou **peso**.

Seja o **vetor de pesos** $\mathbf{w} = \left ( w_1, w_2, \cdots, w_n \right )$; e o **vetor de retornos** $\mathbf{R} = \left ( R_1,R_2, \cdots, R_n \right )$. O retorno do portfolio $R_p$ é definido como sendo o produto escalar entre estes dois vetores, ou seja,

$$ R_p = \mathbf{R} \cdot \mathbf{w} =   \sum_{i=1}^n w_i R_i = w_1 R_1 +  w_2 R_2 + \cdots +  w_n R_n  \ , $$

Onde, naturalmente, $\sum_{i=1}^n w_i = w_1 + w_2 + \cdots + w_n = 1$.

O objetivo da construção de portfolio é justamente encontrar o vetor de pesos $\mathbf{w}$ que produz **o melhor portfolio**, ou seja, o portfolio com a melhor relação risco/retorno, segundo diferentes métricas.

Vamos ver a seguir como trabalhar com este vetor de pesos em python para construirmos nosso portfolio!

```python
df_st.columns
```

**Desvio padrão do portfolio utilizando covariância**

A volativilidade do portfolio pode ser encontrada conforme a seguinte equação:

$$\sigma_{portfolio} = \sqrt{\mathbf{w^T} \Sigma \mathbf{w}}$$

$$ \Sigma (x, y) = \sum_{i=1}^{n}\sum_{j=1}^{n}\frac{1}{n} \left ( (x-\overline{x})^{2} (y-\overline{y})^{2} \right ) $$

- $\sigma_{portfolio}$: volatilidade do portfolio
- $\Sigma$: matrix de covariância dos retornos
- $\mathbf{w}$: vetor de pesos do portfolio

A correlação é medida normalizada da matriz de covariância, que mede a variabilidade conjunta de 2 variáveis aleatórias.

A definição acima é muito importante pois ela é utilizada na **construção de portfolios**, isto é, na escolha dos ativos e dos pesos.

## 4) Capital Asset Pricing Model

O Capital Asset Pricing Model -- CAPM (**Modelo de Precificação de Ativos Financeiros**), é utilizado para determinar a taxa de retorno apropriada de um determinado ativo em relação a uma **carteira de mercado perfeitamente diversificada**, que geralmente é tomada como sendo um índice (IBOV, S&P500, etc.)

No contexto do CAPM, um dos principais conceitos é o **excess return $R_{ex}$**, que é definido como a diferença entre o retorno $R$ e um **retorno livre de risco $R_{RF}$**, isto é, o retorno de um ativo não volátil e extremamente seguro: 

$$R_{ex} = R - R_{RF}$$

A taxa livre de risco é comumente tomada como a taxa de juros do país (no caso do Brasil, é a SELIC), ou as taxas de retorno do tesouro nacional.

O CAPM analisa o risco e retorno de cada ativo, e seu objetivo é mensurar o **risco não diversificável** (risco sistêmico)

$$\mathbb{E}(R_P) - R_{RF} = \beta_p \left (\mathbb{E}(R_M) - R_{RF} \right )$$

- $\mathbb{E}(R_P) - R_{RF}$: retorno esperado em excesso de uma ação ou portfolio $P$;
- $\mathbb{E}(R_P) - R_{RF}$: retorno esperado em excesso de algum portfolio de mercado $M$;
- $\beta_P$: beta do portfolio, ou **exposição**, ao amplo portfólio de mercado $M$.

O beta é então um número que relaciona os retornos de ambos os portfolios. Ao encontrarmos esta constante, é possivel relacionar o seru portfolio com um benchmark de mercado, e assim estimar o risco não diversificável. 

Há algumas formas de calcular o beta. Aqui, veremos o **cálculo do beta pela covariância**, que é dado por:

$$\beta_P = \frac{\Sigma(R_P, R_M)}{\sigma^2(R_M)}$$

- $\Sigma(R_P, R_B)$: covariância entre o portfolio $P$ e o benchmark índice de mercado $M$;
- $\sigma^2(R_B)$: variância do benchmark índice de mercado.

Vamos agora calcular o beta em Python! 

Primeiramente, precisamos ler qual será nosso benchmark de mercado -- como estamos trabalhando com ações brasileiras, usaremos o [IBOVESPA](http://www.b3.com.br/pt_br/market-data-e-indices/indices/indices-amplos/ibovespa.htm).
